## Box Plot

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Define the data for box plot
data = [np.random.normal(0,std,100) for std in range(1,4)]
print(data)

```

```python
len(data)
```

```python
print(data[0].max())
print(data[0].min())
print(data[0].mean())
```

```python
print(data[1].max())
print(data[1].min())
print(data[1].mean())
```

```python
print(data[2].max())
print(data[2].min())
print(data[2].mean())
```

```python
fig, ax = plt.subplots()

# Generate the box plot
ax.boxplot(data)

#Customize the chart
ax.set_title('Box Plot',fontsize=16, fontweight='bold')
ax.set_xticklabels(['Data 1','Data 2', 'Data 3'])
ax.set_xlabel('Data', fontsize=14)
ax.set_ylabel('Value', fontsize=14)
plt.show()
```
