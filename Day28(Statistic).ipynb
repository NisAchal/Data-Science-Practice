```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
#Generate rand num from skewed normal distribution
x = np.random.standard_t(10, size=1000)
x[:15]
```

```python
plt.hist(x, color = 'teal')
```

## MEAN

```python
mean = np.mean(x)
print("Mean:", mean)
```

```python
plt.hist(x, color = 'teal')
plt.axvline(x = mean, color = 'black')
```

## Median

```python
median = np.median(x)
print("Median:", median)
```

```python
plt.hist(x, color = 'teal')
plt.axvline(x = median, color = 'black')
```

## Mode

```python
mode = stats.mode(x).mode
print("Mode:", mode)
```

```python
plt.hist(x, color = 'teal')
plt.axvline(x = mode, color = 'black')
```
