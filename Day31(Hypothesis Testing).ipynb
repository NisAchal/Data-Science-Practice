## Hypothesis testing

```python
import numpy as np 
import matplotlib.pyplot as plt
from scipy import stats
import pandas as pd
```

### Plotting the Rejection Region

```python
#Generate X values for the standard normal distribution
x = np.linspace(-4, 4, 1000)
significance_level = 0.05

#Claculate the critical z-scores
z_critical = stats.norm.ppf(1 - significance_level/2) # ppf = percent point function
print(z_critical)

# plot the standard normal distribution
plt.figure(figsize=(8,6))
plt.plot(x, stats.norm.pdf(x),'black', lw=3, label='Standard Normal Distribution') # pdf = percent distribution function

# shade the rejection region
plt.fill_between(x, stats.norm.pdf(x), where=(x<= -z_critical) | (x>= z_critical), color='red', alpha = 0.5, label = 'Rejection Region')

# Add labels and titles
plt.title('Standard Normal Distribution with Rejection Region') # this shows the title at the top of grapg
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend() #this shows the label on the top right side of graph
plt.tight_layout() # this set the size of graph according to the screen size
plt.show()
```

## Perform Test Statistic and calculate P-Value

```python
# sample data(weight)
sample = [150, 155, 160, 165, 170,175,180 ]

#sample statistics
sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=1)
sample_size = len(sample)

# Hypothesised mean
mu_0 = 165

#clculate the z-score
z_score = (sample_mean - mu_0) / (sample_std / np.sqrt(sample_size)) #z = (x - μ) / (σ / sqrt(n))

# Calculate the P_value
p_value = 2 * (1 - stats.norm.cdf(np.abs(z_score)))
```

```python
p_value
```

#### To print the conclusion

```python
if p_value <= significance_level:
    decision = 'Reject H0'
else:
    decision = 'Fails to reject H0'
print(decision)
```

```python
if p_value <= significance_level:
    conclusion = "There is evidence to support that mean weight is different from 165"
else:
    conclusion = "There is insufficient evidence to support that mean weight is different from 165"
print(conclusion)
```
