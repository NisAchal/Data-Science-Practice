# Measure of dispersion

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
#Generate rand num from skewed normal distribution
x = np.random.standard_t(10, size=1000)
x[:15]
```

```python
plt.hist(x, color='teal')
```

## Variance

```python
variance = np.var(x)
print("Variance:", variance)
```

```python
plt.hist(x, color = 'teal')
plt.axvline(x = x.mean(), color = 'black')
plt.axvline(x = variance, color = 'blue')
```

## Standard Deviation 

```python
std = np.std(x)
print("Standard Deviation:", std)
```

```python
plt.hist(x, color = 'teal')
plt.axvline(x = x.mean(), color = 'black')
plt.axvline(x = std, color = 'blue')
```
